## Creating a new column

```python
df['single_transaction'] = df['Member_number'].astype(str)+'_'+df['Date'].astype(str)
df.head(5)

#The “single_transaction” variable combines the Member_number and date and tells us the item purchased in one receipt.
```

## EXPERIMENT

## 1. Creating the basket

```python
df2 = pd.crosstab(df['single_transaction'], df['itemDescription'])
df2.head()
```

## 2. Encoding the Data

```python
def encode(item_freq):
    res = 0
    if item_freq > 0:
        res = 1
    return res
    
basket_input = df2.applymap(encode)
basket_input
```

## 3. Filter the data

```python
#filtering the data:
data_filter=basket_input[(basket_input>0).sum(axis=1)>=2]
data_filter
```

## 4. Applying the Apriori Algorithm and finding the association

```python
frequent_itemsets = apriori(basket_input, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift")
rules
```

```python
# To get the most frequent item combinations in the entire dataset, sort the dataset by support, confidence and lift
rules.sort_values(["support", "confidence", "lift"], axis = 0, ascending = False).head(10)
```
